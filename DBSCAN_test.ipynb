```python
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Example geo-coordinates data (latitude, longitude)
geo_coordinates = np.array([
    [37.7749, -122.4194],  # San Francisco
    [37.7849, -122.4094],  # San Francisco close by
    [34.0522, -118.2437],  # Los Angeles
    [40.7128, -74.0060],   # New York
    [40.7138, -74.0070],   # New York close by
    [40.7158, -74.0080],   # New York close by
    [37.338331, -121.887377], # Test location
    [37.338331, -121.887377], # Test location
    [37.338173, -121.886750], # Test location
    [37.338310, -121.887657] # Test location
])

# Convert to radians
geo_coordinates_rad = np.radians(geo_coordinates)

# DBSCAN model using Haversine distance
dbscan_model = DBSCAN(eps=0.001,  # roughly 100 meters
                      min_samples=2,  # Minimum points in a cluster
                      metric='haversine')

# Fit the model
clusters = dbscan_model.fit_predict(geo_coordinates_rad)

# Plot the results
fig, ax = plt.subplots()
ax.scatter(geo_coordinates[:, 1], geo_coordinates[:, 0], c=clusters, cmap='rainbow')

for i, txt in enumerate(clusters):
    ax.annotate(txt, (geo_coordinates[:, 1][i]+0.09, geo_coordinates[:, 0][i]+0.09))
```

```python
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Example geo-coordinates data (latitude, longitude)
geo_coordinates = np.array([
    [37.7749, -122.4194],  # San Francisco
    [37.7849, -122.4094],  # San Francisco close by
    [34.0522, -118.2437],  # Los Angeles
    [40.7128, -74.0060],   # New York
    [40.7138, -74.0070],   # New York close by
    [40.7158, -74.0080],   # New York close by
    [37.338331, -121.887377], # Test location
])

# Convert to radians
geo_coordinates_rad = np.radians(geo_coordinates)

# DBSCAN model using Haversine distance
dbscan_model = DBSCAN(eps=0.001,  # roughly 100 meters
                      min_samples=2,  # Minimum points in a cluster
                      metric='haversine')

# Fit the model
clusters = dbscan_model.fit_predict(geo_coordinates_rad)

# Plot the results
plt.scatter(geo_coordinates[:, 1], geo_coordinates[:, 0], c=clusters, cmap='rainbow')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("DBSCAN Clustering of Geo-Coordinates")

plt.show()
```
